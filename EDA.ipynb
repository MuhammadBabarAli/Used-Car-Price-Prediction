```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('pakwheels_used_car_data_v02.csv')
```

```python
# Drop the assembly column due to high number of null values
dataset.drop(columns=['assembly'], inplace=True)
```

```python
dataset.shape
```

```python
# Drop all rows with null values
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
# Get the most common makes of vehicles
common_makes = dataset['make'].value_counts()
print("Most Common Car Makes:")
print(common_makes.head(10))
```

```python
# Get the most common models of vehicles
common_models = dataset['model'].value_counts()
print("\nMost Common Car Models:")
print(common_models.head(10))
```

```python
# Getting the top 5 most popular car colors
top_colors = dataset['color'].value_counts().head(5)
print("The most popular car colors in the dataset are:")
for color, count in top_colors.items():
    print(f"- {color}: {count} vehicles")
```

```python
# Plotting body types against count, to find the most popular body types
# Sedans are the most popular followde by hatchbacks then SUVs by a big difference
plt.figure(figsize=(12, 6))
sns.countplot(dataset, x='body')
plt.xticks(rotation=90)
plt.title('Most Common Body Type')
plt.xlabel('body')
plt.ylabel('count')
plt.tight_layout()
```

```python
# Plotting car body types against price to find out the most expensive car body types
# Off road vehicles are the most expensive followed by SUVs, double cabins, compact SUVs and so on
fig, ax = plt.subplots(figsize=(10, 5))
plt.xticks(rotation=90)
sns.barplot(x="body", y="price", data=dataset, ax=ax)
```

```python
# Plotting vehicle fuel types
# Petrol cars are the most popular
plt.pie(dataset['fuel'].value_counts(), labels=['Petrol', 'Diesel', 'Hybrid'], autopct='%.0f%%')
plt.show()
```

```python
# Plotting fuel types against price to find out the most expensive fuel types
# Diesel cars are the most expensive as they are mostly offroad or SUV vehicles which are the most expensive vehicle types
fig, ax = plt.subplots(figsize=(10, 5))
plt.xticks(rotation=90)
sns.barplot(x="fuel", y="price", data=dataset, ax=ax)
```

```python
# Plotting cities against count, to find the top 10 cities with the most car listings
# Lahore has the most car listings, followed by Karachi, Islamabad and so on
city_ads = dataset['city'].value_counts().reset_index().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='city', y='count', data=city_ads)
plt.title('Car Listings by City (Top 10)')
plt.xlabel('City')
plt.ylabel('Count')
plt.tight_layout()
```

```python
# Plotting cities against average car price, to find the top 10 cities with the most expensive used cars
# None of the cities in the above graph with the most listings are in this graph, most cities are underdeveloped rural type areas with small popultions
prices_avg_city = dataset.groupby(['city'])['price'].mean().reset_index().sort_values(by='price', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='city', y='price', data=prices_avg_city)
plt.xticks(rotation=90)
plt.title('Top 10 High Average Prices Cities')
plt.tight_layout()
```

```python
# Plotting car makes by average price, to find the car makers with the most expensive cars
# Bentley has the most expensive cars followed by GMC, Range(Land Rover), lexus and so on
make_avg_price = dataset.groupby('make')['price'].mean()

plt.figure(figsize=(12, 6))
make_avg_price.plot(kind='bar')
plt.xlabel('Car Make')
plt.ylabel('Average Price')
plt.title('Average Price by Car Make')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
# Plotting the average mileage of vehicles by their body types, to find the most used vehicles
# High roofs have the highest average mileage followed by single cabin, station wagon, trucks. All utilitarian vehicles
mileage_body_avg = dataset.groupby(['body'])['mileage'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(x='body', y='mileage', data=mileage_body_avg)
plt.xticks(rotation=90)
plt.title('Average Mileage by Car Body Type')
plt.xlabel('Body Type')
plt.ylabel('Average Mileage')
plt.tight_layout()
```

```python
# Finding most important features relative to target Price
# Engine size is the most important numerical feature relative to price
corr = dataset[['engine', 'year', 'mileage', 'price']].corr()
corr.sort_values(["price"], ascending = False, inplace = True)
print(corr.price)
```

```python
# Plotting the correlation matrix between numeric features
plt.figure(figsize=(10, 8))
sns.heatmap(dataset[['year', 'mileage', 'price', 'engine']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plotting the distribution of price
# 
plt.figure(figsize=(10, 6))
sns.histplot(dataset['price'], kde=True)
plt.title('Distribution of Price')
plt.show()
```

```python
# Distribution of price shows a high positive skewness, meaning a profusion of outliers
# Applying log transformation to normalize the price distribution
dataset['normalized_price'] = np.log(dataset['price'])
sns.displot(dataset['normalized_price']);
```

```python
# Plotting year against mean car prices, to find the trend between the two
yearly_price_stats = dataset.groupby('year')['price'].mean()

plt.figure(figsize=(12, 6))
plt.plot(yearly_price_stats.index, yearly_price_stats.values, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Mean Car Price (PKR)')
plt.title('Trends in Car Prices Over the Years')
plt.grid(True)
```

```python
# Plotting car prices againt car make, to find the prices of cars offered by each brand
# Bentley has the most expensive cars followed by lexus and so on
plt.figure(figsize=(15, 10))
sns.boxplot(x='make', y='normalized_price', data=dataset)
plt.xlabel('Make', fontsize=14)
plt.ylabel('Price (PKR)', fontsize=14)
plt.title('Price Distribution by Make', fontsize=16)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Plotting price against the vehicle model year
# Slicing the dataset to plot only the first 1000 rows for simplicity
# Newer hybrid cars are more expensive
plt.figure(figsize=(15, 10))
xprop = 'year'
yprop = 'normalized_price'
sns.boxplot(data=dataset.iloc[:1000,:], x=xprop, y=yprop, hue='fuel')
plt.xlabel('{} range'.format(xprop), size=14)
plt.xticks(rotation=90)
plt.ylabel('Number of {}'.format(yprop), size=14)
plt.title('Boxplot of {}(fuel)'.format(yprop), size=20)
plt.show()
```

```python
# Plotting price against the vehicle model year
# Slicing the dataset to plot only the first 1000 rows for simplicity
# Newer automatic cars are more expensive
plt.figure(figsize=(15,10))
xprop = 'year'
yprop = 'normalized_price'
sns.boxplot(data=dataset.iloc[:1000,:], x=xprop, y=yprop, hue='transmission')
plt.xlabel('{} range'.format(xprop), size=14)
plt.xticks(rotation=90)
plt.ylabel('Number of {}'.format(yprop), size=14)
plt.title('Boxplot of {}(transmission)'.format(yprop), size=20)
plt.show()
```

```python
# Plotting car transmission type sales against year, to compare which transmission type sold how much each year
df_trans = dataset.groupby(['year', 'transmission'])[['price']].sum().reset_index()

plt.figure(figsize=(15,6))
sns.barplot(x=df_trans['year'], y=df_trans['price'], hue=df_trans['transmission'], errorbar=None)
plt.title("Transmission Sales according to Year")
plt.ylabel("Sales")
plt.xticks(rotation=45, fontsize = 9)
plt.show()
```

```python
# Plotting car fuel type sales against year, to compare which fuel types sold how much each year
df_fuel = dataset.groupby(['year', 'fuel'])[['price']].sum().reset_index()

plt.figure(figsize=(15,6))
sns.barplot(x=df_fuel['year'], y=df_fuel['price'], hue=df_fuel['fuel'], errorbar=None)
plt.title("Fuel Sales according to Year")
plt.ylabel("Sales")
plt.xticks(rotation=45, fontsize = 9)
plt.show()
```
